```python
import numpy as np 
import pandas as pd 
```

```python
bank = pd.read_csv('bankruptcy.csv')
bank.info()
```

```python
column_names = list(bank.columns)
predictors = column_names[1:96]
target = column_names[0]
```

```python
target
```

```python
X0 = bank[predictors]
```

```python
y0 = bank[target]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train0, X_test, y_train0, y_test = train_test_split(X0, y0, test_size = 0.25, random_state = 50)
```

```python
train = pd.concat([X_train0, y_train0], axis=1)
train
```

```python
test = pd.concat([X_test, y_test], axis=1)
test
```

```python
X = train[predictors]
y = train[target]
X_fv = test[predictors]
y_fv = test[target]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2,random_state =540)
```

```python
y_val
```

```python
y_train
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
pred_val_test = model.predict(X_val)
```

```python
pred_val_test
```

```python
accuracy_score(y_val, pred_val_test, normalize=True, sample_weight=None)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
print("MAE:",mean_absolute_error(y_val, pred_val_test))
print("Accuracy:",accuracy_score(y_val, pred_val_test, normalize=True, sample_weight=None))
print("Classification Report:",'\n',classification_report(y_val,pred_val_test ))

print("Confusion Matrix:", '\n', confusion_matrix(y_val, pred_val_test))
```

```python
model2= DecisionTreeClassifier(max_leaf_nodes = 22)
model2.fit(X_train, y_train)
pred_val_test2 = model2.predict(X_val)
pred_val_train2 = model2.predict(X_train)
print("Accuracy from 40% test data:",accuracy_score(y_val, pred_val_test2, normalize=True, sample_weight=None))
print("Accuracy from 60% train data:",accuracy_score(y_train, pred_val_train2, normalize=True, sample_weight=None))
```
