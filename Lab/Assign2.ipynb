```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('players_subset.csv')
```

```python
data
```

```python
features=['Acceleration','Agility','Balance','Ball control','Composure','Crossing','Curve','Dribbling','Finishing','Free kick accuracy']
```

```python
X = data [features]
#Take the assigned features
```

```python
from sklearn.cluster import KMeans
#Import K means
```

```python
kmeansmodel = KMeans(n_clusters=3)
```

```python
kmeansmodel.fit(X)
```

```python
centroids = kmeansmodel.cluster_centers_
centroids
```

```python
interation = kmeansmodel.n_iter_
interation
```

```python
SSD = kmeansmodel.inertia_
SSD
```

```python
distance = kmeansmodel.transform(X)
distance
#Choose the shortest to one of the centroids
```

```python
labels = kmeansmodel.predict(X)
labels
#Categorize into 3 labels 
```
