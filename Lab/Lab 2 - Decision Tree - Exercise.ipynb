```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train_clean.csv',index_col='PassengerId')
```

```python
train.head()
```

```python
train.info()
```

```python
# select features
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q','Embarked_S']
```

```python
# select target variable
target = ['Survived']
```

```python
# define the datafram for feature and target
X = train [features]
y = train [target]
```

```python
X
```

```python
# import split
from sklearn.model_selection import train_test_split #2 splits
```

```python
# split trainning set and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,random_state =60)
```

```python
X_train
```

```python
y_train
```

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
# fit the model with training data
model.fit(X_train, y_train)
```

```python
# test model with testing data
pred_val = model.predict(X_test)
pred_val
```

```python
y_test
```

```python
# import evaluation metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

## evaluate result 
print("MAE:",mean_absolute_error(y_test, pred_val))
print("Accuracy:",accuracy_score(y_test, pred_val, normalize=True, sample_weight=None))
print("Classification Report:",'\n',classification_report(y_test, pred_val))
print("Confusion Matrix:", '\n', confusion_matrix(y_test, pred_val))

```python
# create tree model object
treeObj = model.tree_
```
